```python
# Daniel Nieves NL = 41
# Algoritmo para cálculo de costo de lápices
# si son mas de 1000 lapices
x = int(input("Ingrese la cantidad de lápices: "))
if x >= 1000:
    pag = x * 0.85
else:
    # Rama No (Bloque naranja)
    pag = x * 0.90

print(f"El pago total es: ${pag:.2f}")

print("Fin del programa")
print("programa creado por Daniel Nieves NL=41")
```
